### 13. Sort categories with the highest profit margins display in bar plot?

Profit Margin=(Total Sales/Total Profit)×100

```python
profit_margin_by_category = (df.groupby('category')['Total Profit'].sum() / df.groupby('category')['Total Sales'].sum()) * 100
```

```python
sorted_profit_margin = profit_margin_by_category.sort_values(ascending=False)
```

```python
plt.figure(figsize = (10,6))
sorted_profit_margin.plot(kind='bar',color='skyblue')
plt.title('Categories with the highest profit margins')
plt.xlabel('Category')
plt.ylabel('Profit Margin (%)')
plt.xticks(rotation=45)
plt.show()
```
