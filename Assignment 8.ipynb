### 5. Are there more male or female customers, Show in Pie Chart?

```python
df.head()
```

```python
df.gender.unique()
```

```python
gender_counts = df.gender.value_counts()
```

```python
plt.figure(figsize=(8, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff'])
plt.axis('equal')  
plt.title('Male vs Female Customers')
plt.show()
```

### 6. Which category is least popular among customers?

```python
df.head()
```

```python
category_counts = df.category.value_counts()
```

```python
least_popular_category = category_counts.idxmin()
least_popular_count = category_counts.min()
print(f"The least popular catgory is'{least_popular_category}' with {least_popular_count} purchases.")
```

### 7. Which category has the highest sales volume?

```python
df.head()
```

```python
df.category.value_counts()
```

```python
higest_sales_volume = df['category'].value_counts()
```

```python
highest_sales_category = higest_sales_volume.idxmax()
highest_sales_count = higest_sales_volume.max()
print(f"The highest sales catgory is'{highest_sales_category}' with {highest_sales_count} purchases.")
```
