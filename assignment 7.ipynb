```python
def categorize_experience(years):
    if years < 5:
        return "< 5 Years"
    elif years < 10:
        return "5-10 Years"
    elif years < 15:
        return "10-15 Years"
    elif years < 20:
        return "15-20 Years"
    elif years < 25:
        return "20-25 Years"
    else:
        return "25+ Years"

df['experience_category'] = df['experience_coding'].apply(categorize_experience)

df
```

```python
compensation_by_category = {
    "< 5 Years": [],
    "5-10 Years": [],
    "10-15 Years": [],
    "15-20 Years": [],
    "20-25 Years": [],
    "25+ Years": []
}

for category, group in df.groupby('experience_category'):
    compensation_by_category[category] = group['compensation'].tolist()

lessthan5 = compensation_by_category["< 5 Years"]
fiveto10 = compensation_by_category["5-10 Years"]
tento15 = compensation_by_category["10-15 Years"]
fifteento20 = compensation_by_category["15-20 Years"]
twentyto25 = compensation_by_category["20-25 Years"]
moretha25 = compensation_by_category["25+ Years"]

lessthan5
fiveto10
tento15
fifteento20
twentyto25
moretha25
```

```python
def categorize_experience(years):
    if years < 5:
        return "< 5 Years"
    elif years < 10:
        return "5-10 Years"
    elif years < 15:
        return "10-15 Years"
    elif years < 20:
        return "15-20 Years"
    elif years < 25:
        return "20-25 Years"
    else:
        return "25+ Years"

df['experience_category'] = df['experience_coding'].apply(categorize_experience)

average_compensation = df.groupby('experience_category')['compensation'].mean()

for category, avg_comp in average_compensation.items():
    print(f"{category}: {avg_comp}")
```
